```python
import math
import numpy as np
import numpy.random as npr
from pylab import plt, mpl

I = 10000 
S0 = 100
M = 50 
T = 2
sigma = 0.25
r = 0.05

ST2 = S0 * npr.lognormal((r - 0.5 * sigma ** 2) * T,
                         sigma * math.sqrt(T), size=I) 

plt.figure(figsize=(10, 6))
plt.hist(ST2, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency');
```
